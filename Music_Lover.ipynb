```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import time

# Replace with your actual YouTube Data API key
developer_key = 'AIzaSyAOCs7as0LN3HIXGTpCZR4Oj1ivLw93eek'
youtube = build('youtube', 'v3', developerKey=developer_key)

# List of music channels' IDs
music_channels = [
    'UCq-Fj5jknLsUf-MWSy4_brA', 'UCbTLwN10NoCU4WDzLf1JMOA', 'UCFFbwnve3yF62-tVXkTyHqg',
    'UCX52tYZiEh_mHoFja3Veciw', 'UCJrDMFOdv1I2k8n9oK_V21w', 'UC56gTxNs4f9xZ7Pa2i5xNzg',
    'UC3MLnJtqc_phABBriLRhtgQ', 'UC_A7K2dXFsTMAciGmnNxy-Q', 'UCnrG75VRwdlp2wtwfpOCBRQ',
    'UCUQg_UBQfVjptn7Wqcgzz-w', 'UCOsyDsO5tIt-VZ1iwjdQmew', 'UCjuV-yJ3ceiHI4ixglXxksA',
    'UC783dnzJqf2ghHp_pFLYbGA', 'UCsaXTlOmt0o9aGeC_n_r8VQ', 'UCtGbExCzlwmsyWKpxLnyEww',
    'UCcvNYxWXR_5TjVK7cSCdW-g', 'UColNRmGBE3dNzJ89zZZOh2Q', 'UClZF7OsrECe3t_ADrjZEh_Q',
    'UCopY0NAzASqmD6eVaRj2TFQ', 'UCF8Ar7alYrtOI5SF6b7T0xA', 'UCQjHoPr0kqN83Mmihy7nsWw'
]

def get_channel_videos(youtube, channel_id):
    video_ids = []
    playlist_id = get_channel_playlist_id(youtube, channel_id)

    next_page_token = None
    while True:
        try:
            response = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=playlist_id,
                maxResults=50,
                pageToken=next_page_token
            ).execute()

            video_ids.extend([item['contentDetails']['videoId'] for item in response.get('items', [])])
            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break

        except HttpError as e:
            if e.resp.status == 403 and 'quotaExceeded' in str(e):
                print("Quota exceeded. Waiting for a minute before retrying...")
                time.sleep(60)
                continue
            else:
                print(f"An HTTP error occurred: {e}")
                break  # Handle other HTTP errors gracefully or retry

    video_details = []
    for i in range(0, len(video_ids), 50):
        response = youtube.videos().list(
            part='snippet,statistics',
            id=','.join(video_ids[i:i + 50])
        ).execute()

        video_details.extend([
            {
                'Title': video['snippet']['title'],
                'Video_URL': f'https://www.youtube.com/watch?v={video["id"]}',
                'Views': video['statistics'].get('viewCount', 0)
            } for video in response.get('items', [])
        ])

    return video_details

def get_channel_playlist_id(youtube, channel_id):
    channel_response = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    ).execute()
    return channel_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']

# Main code for top songs
all_top_videos = []

for channel_id in music_channels:
    channel_videos = get_channel_videos(youtube, channel_id)
    time.sleep(1)  # Sleep to avoid reaching API quota limits
    top_videos = [video for video in channel_videos if int(video['Views']) > 100000000]
    all_top_videos.extend(top_videos)

# Create a pandas DataFrame and save as a CSV file
if all_top_videos:
    pd.DataFrame(all_top_videos).to_csv('top_songs.csv', index=False)
    print("Top songs details saved to top_songs.csv")
else:
    print("No top songs details to save.")
```
